```python
from ultralytics import YOLO
import os

dataset_path = "/content/ALPR-YOLOv8-4"
data_yaml = os.path.join(dataset_path, "data.yaml")

model = YOLO("yolov8s.pt")  # أو yolov8m.pt إذا GPU يسمح

# تدريب النموذج
model.train(
    data=data_yaml,
    epochs=50,
    imgsz=640,
    batch=16,
    lr0=0.001,
    augment=True,
    optimizer="AdamW",
    name="alpr_yolov8s_train",  # اسم المشروع
    project="ALPR_Project",      # مشروع W&B
    exist_ok=True,
    patience=10                  # early stopping إذا لم يتحسن val mAP
)
```
